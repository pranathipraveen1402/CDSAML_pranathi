```python
import pandas as pd
import numpy as np
df = pd.read_csv('/kaggle/input/iris-dataset/iris.data.csv',header = None)
df
```

```python
df.columns = ['sepallength','sepalwidth','petallength','petalwidth','class']
df
```

```python
X = df.drop(['class'],axis = 1)
y = df['class']
X
```

```python
# scaling to common range - brings all columns in common range
#from sklearn.preprocessing import MinMaxScaler

#scalar = MinMaxScaler()
#X_scaled = scalar.fit_transform(X)
#X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size = 0.4,random_state=42)
# 30 test,120 train
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
# fit data to gnb object by giving training and testing samples
gnb.fit(X_train,y_train)

# use predict function
y_pred = gnb.predict(X_test) #y_pred contains predictions and y_test contains original labels
```

```python
y_pred
```

```python
y_test
```

```python
import sklearn.metrics

lbs = ['Iris-virginica','Iris-setosa','Iris-virginica']
print(sklearn.metrics.confusion_matrix(y_test,y_pred,labels = lbs))
print(sklearn.metrics.accuracy_score(y_test,y_pred))
print(sklearn.metrics.precision_score(y_test,y_pred,labels = lbs,pos_label=1,average='micro'))
print(sklearn.metrics.recall_score(y_test,y_pred,labels = lbs,pos_label=1,average='micro'))
print(sklearn.metrics.f1_score(y_test,y_pred,labels = lbs,pos_label=1,average='micro'))
```
